```python
# np.array
x = [1, 2, 3]
y = np.array(x)
print(y) 

z = np.array(y)
print(z)  #  but `z` is a new array, not the same object as `y`
```

```python
# np.asarray
a = np.array([1, 2, 3])
b = np.asarray(a)
print(a is b) 
c = np.asarray([1, 2, 3])
print(c is a) 
```

```python
# np.asanyarray
a = np.array([1, 2, 3])
b = np.asanyarray(a)
print(a is b) 

m = np.matrix([[1, 2], [3, 4]])
n = np.asanyarray(m)
print(type(n))
```

```python
# shallow copy 
import copy
original = [1, [2, 3], 4]
shallow_copy = copy.copy(original)
print(shallow_copy is original)  
print(shallow_copy[1] is original[1])  
```

```python
#deep copy 
import copy
original = [1, [2, 3], 4]
deep_copy = copy.deepcopy(original)
print(deep_copy is original)  
print(deep_copy[1] is original[1])  
```

# 14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

# b) Label the x-axis as 'Date'.

# c) Set the title of the plot as 'Temperature and Humidity Over Time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a sample time-series dataset
# Generate a date range
dates = pd.date_range(start='2024-01-01', periods=100, freq='D')

# Generate random temperature and humidity values
temperature = np.random.uniform(low=-10, high=35, size=len(dates))
humidity = np.random.uniform(low=20, high=100, size=len(dates))

# Create a DataFrame
data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set 'Date' as the index
data.set_index('Date', inplace=True)

# a)  Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot temperature data
ax1.plot(data.index, data['Temperature'], color='red', label='Temperature')

# b) Lable x axix as Date
ax1.set_xlabel('Date')


ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

#  Create a second y-axis for the humidity data
ax2 = ax1.twinx()
ax2.plot(data.index, data['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

#  Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Add a legend
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

fig.autofmt_xdate()  # Rotate the x-axis labels

# Step 6: Display the plot
plt.show()
```

# 15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

# a) Plot a histogram of the data with 30 bins.

# b) Overlay a line plot representing the normal distribution's probability density function (PDF).

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

# d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Step 1: Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc =0 , scale =1, size = 1000)


# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay a line plot representing the normal distribution's PDF

x = np.linspace(max(data),min(data), 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# c) Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

#  Add a legend
plt.legend()

# display 
plt.show()
```

# 17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create the scatter plot with Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='deep')

# Step 5: Add a legend, label the axes, and set the title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Step 6: Show the plot
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--') # for modification 
plt.show()

```

# 18  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

#  Generate the data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)  # 0 to 4π
y = np.sin(x)

# Create a Bokeh plot
output_notebook()  # This is for displaying the plot in a Jupyter notebook
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis', width=800, height=400)

#  Add a line to the plot
p.line(x, y, legend_label='Sine Wave', line_width=2)

#  Customize the plot (grid lines are added by default in Bokeh)
p.legend.location = "top_left"

#  Display the plot
show(p)
```

# 19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import random

#  Generate the data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a Bokeh plot
source = ColumnDataSource(data=dict(categories=categories, values=values))


#creating a bokeh figure 
# Giving the title , height ,lable
p = figure(x_range=categories, height=400, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values', tools='hover', tooltips=[('Value', '@values')])

# Add bars to the plot
p.vbar(x='categories', top='values', width=0.5, source=source, 
       line_color='white', fill_color='colors')

# Customize the plot (color bars based on values)
colors = ['#FF5733', '#FFC300', '#C70039', '#900C3F', '#581845']
source.data['colors'] = [colors[random.randint(0, len(colors) - 1)] for _ in range(len(categories))]

# Display the plot
show(p)
```
