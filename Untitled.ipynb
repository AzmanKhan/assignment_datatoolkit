```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

#  Generate the data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)  # 0 to 4Ï€
y = np.sin(x)

# Create a Bokeh plot
output_notebook()  # This is for displaying the plot in a Jupyter notebook
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis', width=800, height=400)

#  Add a line to the plot
p.line(x, y, legend_label='Sine Wave', line_width=2)

#  Customize the plot (grid lines are added by default in Bokeh)
#p.legend.location = "top_left"

#  Display the plot
show(p)
```
